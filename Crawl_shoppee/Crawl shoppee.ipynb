```python
import pandas as pd

data_iphone = 'D:/crawdata/Crawl_shoppee/iphone.xlsx'
iphone = pd.read_excel(data_iphone)

iphone['Price'] = iphone['Price'].str.replace('.', '')
iphone['Sold'] = iphone['Sold'].str.replace('Đã bán ', '')
iphone['Discount'] = iphone['Discount'].str.replace('-', '')
iphone['Discount']  = iphone['Discount'].str.replace('%', '')
iphone.fillna('0', inplace=True)
```

```python
sold_iphone = []
for item in iphone['Sold']:
    if 'k' in item:
        if ',' in item:
            item = item.replace(',', '')
            item = item.replace('k', '00')
            sold_iphone.append(item)
        else:
            item = item.replace('k', '000')
            sold_iphone.append(item)
    else:
        sold_iphone.append(item)
        
iphone['Sold'] = sold_iphone

print(iphone)
```

```python
iphone['Price'] = iphone['Price'].astype(int)
iphone['Discount'] = iphone['Discount'].astype(int)
iphone['Sold'] = iphone['Sold'].astype(int)

iphone.info()
```

```python
iphone.describe()
```

```python
import numpy as np
x = np.array(['iphone','macbook','ipad','apple watch', 'air_pord', 'accessories'])
y = np.array([len(iphone['Categories']),len(macbook['Categories']),len(ipad['Categories']),len(apple_watch['Categories']),len(air_pord['Categories']), len(accessories['Categories'])])

plt.pie(y, labels=x, autopct='%1.2f%%')
plt.title('Các loại mặt hàng shop bán')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["Categories"], df["Sold"], color='skyblue')
plt.xlabel('Mặt hàng')
plt.ylabel('Số lượng đã bán')
plt.title('Biểu đồ số lượng sản phẩm đã bán cho mỗi mặt hàng')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Revenue'] = df['Price'] * df['Sold']
revenue_by_category = df.groupby('Categories')['Revenue'].sum()
print(revenue_by_category)
```

```python
plt.figure(figsize=(10, 6))
revenue_by_category.plot(kind='bar', color='skyblue')
plt.xlabel('Mặt hàng')
plt.ylabel('Mặt hàng')
plt.title('Biểu đồ doanh thu theo danh mục')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
filtered_df = df[df['Discount'] > 0]

grouped_df = filtered_df.groupby('Categories')['Discount'].count()

plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', color='skyblue')
plt.xlabel('Mặt hàng')
plt.ylabel('Tổng ưu đãi')
plt.title('Biểu đồ tổng ưu đãi theo danh mục sản phẩm')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker

plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Mức giá(triệu đồng)')
plt.ylabel('Tần suất')
plt.title('Histogram của giá sản phẩm')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: '{:,.0f}'.format(x/1000000)))
plt.grid(True)
plt.show()
```
